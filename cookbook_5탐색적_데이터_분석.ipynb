```python
import seaborn as sns
fig, ax = plt.subplots(figsize = (8,8))
corr = fueleco[["city08", "highway08", "cylinders"]].corr()
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
sns.heatmap(
    corr,
    mask = mask,
    fmt = ".2f",
    annot = True,
    ax = ax,
    cmap = "RdBu",
    vmin = -1,
    vmax = 1,
    square = True,
)
```
